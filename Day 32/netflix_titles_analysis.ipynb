```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")

netflix_data = pd.read_csv("netflix_titles.csv")
df = pd.DataFrame(netflix_data)
df.head()
```

```python
df.duplicated()
df.drop_duplicates()
# netflix_data.drop_duplicates()
```

```python
Mean = df.mean(numeric_only=True)
df.fillna(Mean)
```

```python
total_movies = df["type"].value_counts()["Movie"]
total_TV_Show = df["type"].value_counts()["TV Show"]
print(f"Total Movies : {total_movies}")
print(f"Total TV Shows : {total_TV_Show}")
```

```python
print(df["country"].value_counts().head(20))
```

```python
print("Most Common Rating : ", df["rating"].value_counts().idxmax())
print("Count : ", df["rating"].value_counts().max())
# print(df["rating"].value_counts().head(1))
```

```python
titles_per_year = df['release_year'].value_counts().sort_index()
plt.figure(figsize=(12, 6))
titles_per_year.plot(kind='bar', color='skyblue')
plt.title('Number of Netflix Titles Released Each Year')
plt.xlabel('Release Year')
plt.ylabel('Number of Titles')
plt.show()
```

```python
missing_entries = df[df[['director', 'cast', 'country']].isnull().any(axis=1)]
print("Missing Entries : ", len(missing_entries))
```

```python
#6.	Replace all missing country values with "Unknown". How many are now labeled as "Unknown"?
df['country'].fillna('Unknown')
unknown_count = (df['country'] == 'Unknown').sum()
print("Unknown Count : ", unknown_count)
```

```python
# Filter rows where director is missing but cast is present
filtered = df[df['director'].isnull() & df['cast'].notnull()]
# Count the number of such rows
count = len(filtered)
print("Number of entries with missing director but available cast:", count)
```

```python
# 8. Extract the number of seasons or minutes from the duration column into separate columns (num_minutes or num_seasons).
df['duration_num'] = df['duration'].str.extract('(\d+)')
df['num_minutes'] = df['duration_num'].where(df['type'] == 'Movie')
df['num_seasons'] = df['duration_num'].where(df['type'] == 'TV Show')
print(df[['type', 'num_minutes', 'num_seasons']].head(10))
```

```python
# 9. What is the average duration of movies (in minutes)?
df['num_minutes'] = df['duration'].where(df['type'] == 'Movie').str.extract(r'(\d+)').astype(float)

# Calculate average duration
average_duration = df['num_minutes'].mean()

print("Average duration of movies:", round(average_duration, 2), "minutes")
```

```python
# 10. Among TV Shows, which title has the highest number of seasons?
highest = df["type"] == "TV Show"
print(highest.value_counts().max())
```

```python
# 11. List all movies released in 2020 from India.
movies_2020_india = df[
    (df['type'] == 'Movie') & 
    (df['release_year'] == 2020) &
    (df['country'] == 'India')
]
print("Movies released at year 2020 : \n", movies_2020_india[['title', 'release_year', 'country']])
```
