```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
from sklearn.impute import SimpleImputer
from sklearn.decomposition import PCA

df = pd.read_csv("CC GENERAL.csv")
```

```python
df.drop("CUST_ID", axis=1, inplace=True)
imputer = SimpleImputer(strategy="mean")
df_imputed = imputer.fit_transform(df)
```

```python
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_imputed)
```

```python
from sklearn_extra.cluster import KMedoids

silhouette_scores = []
K = range(2, 11)

for k in K:
    kmedoids = KMedoids(n_clusters=k, random_state=42, method='pam').fit(df_scaled)
    labels = kmedoids.labels_
    sil_score = silhouette_score(df_scaled, labels)
    silhouette_scores.append(sil_score)
```

```python
plt.figure(figsize=(8, 5))
plt.plot(K, silhouette_scores, marker='o')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Method for Optimal k')
plt.grid(True)
plt.xticks(K)
plt.show()
```

```python
optimal_k = K[np.argmax(silhouette_scores)]
print(f"Optimal number of clusters: {optimal_k}")
```

```python
kmedoids_final = KMedoids(n_clusters=optimal_k, random_state=42, method='pam').fit(df_scaled)
final_labels = kmedoids_final.labels_
```

```python
pca = PCA(n_components=2)
df_pca = pca.fit_transform(df_scaled)
```

```python
plt.figure(figsize=(8, 6))
for i in range(optimal_k):
    plt.scatter(df_pca[final_labels == i, 0], df_pca[final_labels == i, 1], label=f'Cluster {i}')

plt.title('K-Medoids Clusters (PCA reduced)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.grid(True)
plt.show()
```
