```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 
sns.set_style("whitegrid")

iris_data = pd.read_csv("Iris.csv")
# First 10 rows 
iris_data.head(10)
```

```python
print("Total rows and columns :", iris_data.shape)
```

```python
print("Column names and data types : \n")
iris_data.info()
```

```python
print("Describing numeric features : ")
iris_data.describe()
```

```python
FilteredData = iris_data[iris_data["SepalLengthCm"] > 3]
print(FilteredData)
```

```python
petallen_and_species = iris_data[["PetalLengthCm", "Species"]]
print("Petal Length and its Species : \n", petallen_and_species)
```

```python
versicolor_species = iris_data[iris_data["Species"] == "Iris-versicolor"]
print(versicolor_species)
```

```python
# sorted_data = np.sort(iris_data[["SepalWidthCm"]])
sorted_data = iris_data.sort_values(by = "SepalWidthCm", ascending = False)
print("Sorted Data according to Sepal Width in descending order :\n", sorted_data)
```

```python
grouped_data = iris_data.groupby("Species")
print(f"Mean : \n{grouped_data.mean()}")
print(f"\nMax : \n{grouped_data.max()}")
print(f"\nCount : \n{grouped_data.count()}")
```

```python
avg_petal_len = grouped_data["PetalLengthCm"].mean()
highest_petal_species = avg_petal_len.idxmax()
highest_petal_len = avg_petal_len.max()

print(f"{highest_petal_species} is the highest petal length species with length {highest_petal_len:.2f}")
```

```python
plt.hist(iris_data["SepalLengthCm"], bins = 10, color = "orange", edgecolor = "red")
plt.title("Histogram of Sepal Length")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.boxplot(x = "Species", y = "PetalWidthCm", data = iris_data)
plt.title("Boxplot of Petal Width by Species.")
plt.xlabel("Species")
plt.ylabel("Petal Width (cm)")
plt.show()
```

```python
sns.scatterplot(
    x="SepalLengthCm",
    y="PetalLengthCm",
    hue="Species",
    data=iris_data,
)

plt.title("Scatter Plot of Sepal Length vs Petal Length")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Petal Length (cm)")
plt.show()
```

```python
sns.pairplot(iris_data, hue = "Species", palette = "Set2")
plt.suptitle("Pairplot of Iris Dataset", y = 1)
plt.show()
```
