```python
mean_temperature = df["Temperature"].mean()
mode_temperature = df["Temperature"].mode()
median_temperature = df["Temperature"].median()
print("Mean temperature : ", mean_temperature)
print("Mode temperature :\n", mode_temperature)
print("Median temperature : ", median_temperature)
temperature_std = df["Temperature"].std()
temperature_var = df["Temperature"].var()
print(f"Standard Deviation of Temperature : {temperature_std:.2f}")
print(f"Variance of Temperature : {temperature_var:.2f}")
```

```python
mean_fuel_price = df["Fuel_Price"].mean()
mode_fuel_price = df["Fuel_Price"].mode()
median_fuel_price = df["Fuel_Price"].median()
print(f"Mean Fuel Price :  {mean_fuel_price:.2f}")
print("Mode Fuel Price :\n", mode_fuel_price)
print("Median Fuel Price : ", median_fuel_price)
fuel_price_std = df["Fuel_Price"].std()
fuel_price_var = df["Fuel_Price"].var()
print(f"Standard Deviation of Fuel Price : {fuel_price_std:.2f}")
print(f"Variance of Fuel Price : {fuel_price_var:.2f}")
```

```python
mean_CPI = df["CPI"].mean()
mode_CPI = df["CPI"].mode()
median_CPI = df["CPI"].median()
print(f"Mean CPI :  {mean_CPI:.2f}")
print(f"Mode CPI :  {mode_CPI}")
print(f"Median CPI : {median_CPI:.2f}")
CPI_std = df["CPI"].std()
CPI_var = df["CPI"].var()
print(f"Standard Deviation of CPI : {CPI_std:.2f}")
print(f"Variance of CPI : {CPI_var:.2f}")
```

```python
mean_unemployment = df["Unemployment"].mean()
mode_unemployment = df["Unemployment"].mode()
median_unemployment = df["Unemployment"].median()
print(f"Mean Unemployment :  {mean_unemployment:.2f}")
print(f"Mode Unemployment :  {mode_unemployment}")
print(f"Median IUnemployment : {median_unemployment:.2f}")
Unemployment_std = df["Unemployment"].std()
Unemployment_var = df["Unemployment"].var()
print(f"Standard Deviation of Unemployment : {Unemployment_std:.2f}")
print(f"Variance of Unemployment : {Unemployment_var:.2f}")
```
